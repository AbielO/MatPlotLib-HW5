```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset

city = city_df.set_index(['city'])
ride = ride_df.set_index(['city'])

combined_df = pd.merge(ride, city, on='city' )

# Display the data table for preview
combined_df
```

```python
city_final = combined_df.groupby(["city"])
total_df = city_final.sum()

```

```python
city_type = combined_df['type']
ride_total = ride_df["city"].value_counts()
fare = total_df['fare']
ave_fare = fare / ride_total
total_df["Total_Rides"] = ride_total
total_df["Average_Fare"] = ave_fare.round(2)

```

```python
final_df = pd.merge(total_df, city, on='city' )
final_df.head()
```

```python
rural = final_df[final_df["type"] == "Rural"]
suburban = final_df[final_df["type"] == "Suburban"]
urban = final_df[final_df["type"] == "Urban"]
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x=list(final_df.Total_Rides)
y=list(final_df.Average_Fare)
s=list(final_df.driver_count_x)


# Build the scatter plots for each city types
plt.scatter(rural["Total_Rides"], rural["Average_Fare"], s = rural["driver_count_x"], color = "gold",
            edgecolor = "black", label = "Rural", alpha = .75)

plt.scatter(suburban["Total_Rides"], suburban["Average_Fare"], s = suburban["driver_count_x"], color = "skyblue",
            edgecolor = "black", label = "Suburban", alpha = .75)

plt.scatter(urban["Total_Rides"], urban["Average_Fare"], s = urban["driver_count_x"], color = "coral",
            edgecolor = "black", label = "Urban", alpha = .75)
# Incorporate the other graph properties
plt.title("Pyber Sharing Data 2016")
#plt.legend(handles=['Rural', 'Suburban', 'Urban'], loc="best")
plt.title("Pybar Ride Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare(s)")
# Create a legend
plt.legend()
plt.grid()

# Incorporate a text label regarding circle size
plt.text(50,30, "Cirlce size correlates with driver count per city.", fontsize = 12)
# Save Figure
plt.savefig("../Images/bubbleplot.png")

plt.show()
```
